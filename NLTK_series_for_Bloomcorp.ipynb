Now what is stopwords?? 
'a', 'the', 'and' these are the words we call stopwords which doesnot mean anything in terms of input to a model.

```
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```
sent = 'my name is abhijit and i am a data scientist.the primary goal of any data science consultant is to know what is the hidden pattern this data holds'

# print(word_tokenize(sent))
words = word_tokenize(sent)
words
```

```
stop_word= set(stopwords.words('english'))
print(stop_word)
```

```
new_sent= []
for word in words:
  if word not in stop_word:
    new_sent.append(word)
print(new_sent)
```

Can you see the removed words?

So now let's talk of "Stemming"

```
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```
sent = 'so here we are reading the book, driving our life, activated so many things'
words = word_tokenize(sent)
print(words)
```

```
stmer = PorterStemmer()
for word in words:
  print(stmer.stem(word))
```

Did you see how it changed the words to the normal form, like driving to drive, reading to read activated to activ..
its not that important as wordnet and other tools are there, it's not that important few times.


So wanna know more, let's see what is 'Part of speech tagging'

```
nltk.download('state_union')
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```
train_sent = state_union.raw('2005-GWBush.txt')
test_sent = state_union.raw('2006-GWBush.txt')
custom_sent_token = PunktSentenceTokenizer(train_sent)
token_sent = custom_sent_token.tokenize(test_sent)
```

```
def text_process():
  for word in token_sent:
    words = nltk.word_tokenize(word)
    tagged_words = nltk.pos_tag(words)
    return tagged_words
```

```
text_process()
```

Let's see lemmatization

```
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
words = ['book', 'booking', 'booked', 'bookish']
for word in words:
  print(lem.lemmatize(word))
```

I think you don't get it, let's grab our old sentence

```
sent = 'so here we are reading the book, driving our life, activated so many things'
words = [w for w in word_tokenize(sent) ]
for word in words:
  print(lem.lemmatize(word))
```

So it is like stemming but works much better with keeping the meanin intact, not like 'driving' to 'driv'

```
import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
word = wordnet.synsets('go')
for w in word:
  print(w)
```

these are the different meaning of go in different usage of "GO" 
we can view the definition as well. how?? well easy


```
word = wordnet.synsets('go')
for w in word:
  print(w,w.definition())
```

can we check any semantic similarity??
Sure..here it is


```
word1 = wordnet.synset('purse.n.01')
word2 = wordnet.synset('wallet.n.01')
print(word1.wup_similarity(word2))
```

So the similarity is 82%
