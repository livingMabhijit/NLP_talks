## We will learn about Word Embedding.. Let's go

#### Word Embedding works as follows:
    vocab size
    one_hot function
    pass that to embeding layer
    initialize the dimensions

```python
from keras.preprocessing.text import one_hot
```

```python
sentences = ['he is a nice guy, I know him since 2008',
             'she is a nice lady, her house is few yards away from mine',
             'would you like to eat apple?',
             'did you eat orange from fridge?',
            'this is the best place to eat in Paris.',
             'munich is a must visit destination',
            "i have not seen this view in my life, of it's so beautiful",
            'amazing scenery,loved it']
```

```python
sentences
```

```python
vocab_size = 5000
```

### One hot representation

```python
onehot_phase = [one_hot(words,vocab_size) for words in sentences]
print(onehot_phase)
```

## Word embedding representation

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
import numpy as np
```

```python
#make all the length same, so using padding
sent_length = 10
embed_doc = pad_sequences(onehot_phase,padding='pre',maxlen=sent_length)
print(embed_doc)
```

```python
dim = 30
model = Sequential()
model.add(Embedding(vocab_size,40,input_length=sent_length))
model.compile('adam','mse')
model.summary()
```

```python
pred = model.predict(embed_doc)
```
